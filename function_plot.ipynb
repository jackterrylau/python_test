```python
import numpy as np

# np.array object sample
print("-------npArray1-------")
npArray1 = np.array([0,2,3,4])
print(npArray1)
print("npArray1.type = %s"%(type(npArray1)))
print("npArray1.data type : %s"%npArray1.dtype)
print("npArray1.(col) :", npArray1.shape)

print("-------npArray2-------")
npArray2 = np.array([
[1,2,4,10],
[5,8,9,8],
[0,0,0,0]    
])
print(npArray2)
print("npArray1.(row, col) :", npArray2.shape)
```

```python
'''
numpy array structure : np array 內部元素類別一致.
np.genfromtxt(file, delimiter, dtype) : 從文件讀取數據.
np.arange(n) : generate an array with n number 0 - (n-1).
np.zeros(row, col) : initialize matrix[row, col] as 0
np.ones(row, col) : initialize  matrix[row, col] as 1
np.linspace(start, end, number): 產生 number 個 遞增數字 從 start to end

from numpy import random as rd
rd.random(row, col) : 產生 row*col 個 0<X<1 的隨機數 形成 Matrix[row ,col]

np.dot(a1, a2) : npArray 相乘
np.floor(npArray) : 取 npArray 的 地板值（向下取整）
np.exp(n) : e 的 n 次方

'''

numpyData = np.genfromtxt('numpyData.txt', delimiter=',', dtype='str')
print("numpyData = ", numpyData)
print("numpyData[0,2] = ", numpyData[0,2])

# numpy array slice - npArray[row_slice, col_slice] npArray[:,:]
print("讀取第1列所有元素: numpyData[0:1] = ", numpyData[0:1])
print("讀取第3行所有元素: numpyData[:,2] = ", numpyData[:,2])
print("讀取第3行所有元素: numpyData[:,0:2] = ", numpyData[:,0:2])
print("numpyData[:,m:n] 從第 m 行 到 第 n 行")

# np.arange(n) 隨機產生 n 個 0 - n-1 的整數
print("產生 0-9 的 numpy 陣列 = ", np.arange(10))
# np.arange(start, end, interval) 隨機產生 n 個 start - end-1 且間隔為 interval 的整數
print("產生 5-20 的 numpy 陣列 間隔值3 = ", np.arange(5, 20, 3))

# 在 1 - pi 之間產生 50 個 隨機數
import math
print("產生 1-pi 之間產生 50 個 遞增的 隨機數 = ", np.linspace(1, math.pi, 50))

```

```python
'''
numpy array 值比較
numpy array & 及 | 運算
'''

# 比對某一 npArray 所有元素 是否 等於 某一個值
npArrayComp1 = np.array([1, 2, 3, 4])
print("比對 npArrayComp1 每一個值是否等於 5: ", npArrayComp1 == 5)
npArrayComp2 = np.array([[1,2,3,4], [5,6,7,8]])
print("比對 npArrayComp2(多維) 每一個值是否等於 3: ", npArrayComp1 == 3)
print("比對 矩陣中某一行所有元素是否為 5: ", npArrayComp2[:,2] == 5)
print("比對 矩陣中某一行所有元素是否為 7: ", npArrayComp2[:,2] == 7)
print("比對 矩陣中某一行所有元素是否為 7, 如果有某一行元素為7 則取出該元素所在列的所有元素: ", npArrayComp2[npArrayComp2[:,2] == 7, :])

# and 及 or operations
npArrayLogic = np.array([[1,2,3,4], [5,6,7,8]])
print("判斷 npArrayLogic 每一個值是否同時能被 2和3 整除: ", (npArrayLogic%2 == 0)&(npArrayLogic%3 == 0))
print("判斷 npArrayLogic 每一個值是否等於 2或3 : ", (npArrayLogic == 2)|(npArrayLogic == 3))
npArrayLogic[(npArrayLogic == 2)|(npArrayLogic == 3)] = 10
print("將 npArrayLogic 每一個值為 2或3 的元素用 10 取代: ", npArrayLogic)
print("將 npArrayLogic 元素為 10 的那一欄所有元素列出: ", npArrayLogic[:,npArrayLogic[0,:]==10])
```

```python
#pandas

import pandas as pd 

idcard_info = pd.read_csv('idcard.csv', encoding = 'utf8')
print(type(idcard_info))
print(idcard_info.dtypes)

# 顯示前五條數據
print("pandas 顯示前五條數據：")
idcard_info.head() # 5 is by default

# 顯示後3條數據
print("pandas 顯示後3條數據：")
print(idcard_info.tail(3))

print("pandas.columns：")
print(idcard_info.columns)

print("pandas 獲取指定某一條數據:")
print(idcard_info.loc[7])

#print("pandas 獲取指定某一欄所有數據:")
#pd_money = idcard_info["money"]
#print(pd_money)

print("pandas 獲取指定某一數字欄位所有數據且乘以 10:")
#pd_10money =idcard_info["money"] * 10
#print(pd_10money)

print("pandas 獲取指定某一數字欄位所有數據的 max/min/mean:")
pd_money_max =idcard_info["money"].max()
pd_money_min =idcard_info["money"].min()
pd_money_mean =idcard_info["money"].mean()
print("pd_money_max = ", pd_money_max)
print("pd_money_min = ", pd_money_min)
print("pd_money_mean = ", pd_money_mean)

print("pandas 用既有欄位算出新的欄位:")
cMoney = idcard_info["confidence"]/idcard_info["money"]
idcard_info["CMoney"] = cMoney
print(idcard_info)
```

```python
# 必須先把 DataFrame Object 轉成 np.array 才能帶入 plt.plot 作圖
confidence_npArray = idcard_info["confidence"].values
money_npArray = idcard_info["money"].values
plt.plot(confidence_npArray, money_npArray)

#針對圖做一些指引
plt.xticks(rotation = -45) # x axis value ratate 45 degrees
plt.yticks(rotation = 45) # y axis value ratate 45 degrees
plt.title("confidence - money") # set draw title
plt.xlabel("confidence") # set x label name
plt.ylabel("money") # set y label name

plt.show()
```

```python
'''
draw multiple plots
'''

fig = plt.figure() # 一個大 Figure 可以放置多張 Plot
sf1 = fig.add_subplot(2, 2, 1) # 一個 子 plot 放在 2*2 Figure 的 第一個 位置
sf2 = fig.add_subplot(2, 2, 4) # 一個 子 plot 放在 2*2 Figure 的 第四個 位置
# Set first subplot 且指定線的顏色
sf1.plot(confidence_npArray, money_npArray, c="red")
# Set second subplot 且指定線的顏色
sf2.plot(money_npArray, confidence_npArray, c="black")

plt.show()
```

```python
# Draw multiple lines in one plot.
import numpy as np

# figsize 代表生成一張圖其 size = 16*6, 相當於 plt.figure(1, 1, figsize=(16,6))
fig = plt.figure(figsize=(16,6)) 
colors = ['red','blue','green','orange','black']
h = np.arange(10, 100, 10) # generate 10 - 100 random integers with step 10 for x
for i in range(5):
    # np.random.randn(9, 1) : 生成 9 欄 1 列 的隨機數
    v = 1000*i + np.random.randn(9, 1)*1000 # generate random integers for y
    label_num = str(i*10)
    plt.plot(h, v, c = colors[i], label = label_num)
plt.legend(loc = 'best') # 給圖加上一個圖例 並且自動放在圖中最佳不會遮擋到線的位置

plt.show()
```
